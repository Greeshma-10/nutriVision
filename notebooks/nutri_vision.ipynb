```python
import os
import random
import shutil

# Adjust these paths to your drive setup
base_dir = "/content/datasets"
train_img_dir = os.path.join(base_dir, "train/images")
train_lbl_dir = os.path.join(base_dir, "train/labels")

valid_img_dir = os.path.join(base_dir, "valid/images")
valid_lbl_dir = os.path.join(base_dir, "valid/labels")

# Create valid folders if they don't exist
os.makedirs(valid_img_dir, exist_ok=True)
os.makedirs(valid_lbl_dir, exist_ok=True)

# List all image files
images = [f for f in os.listdir(train_img_dir) if f.endswith(".jpg") or f.endswith(".png")]
random.shuffle(images)

# Split 80% train, 20% valid
split_index = int(len(images) * 0.8)
valid_images = images[split_index:]

for img_file in valid_images:
    label_file = img_file.replace(".jpg", ".txt").replace(".png", ".txt")

    # Move image
    shutil.move(os.path.join(train_img_dir, img_file), os.path.join(valid_img_dir, img_file))

    # Move corresponding label
    shutil.move(os.path.join(train_lbl_dir, label_file), os.path.join(valid_lbl_dir, label_file))

print(f"✅ Moved {len(valid_images)} images + labels to valid/")
```

```python
import urllib.request
import os

# ✅ Create a folder to save the images
save_dir = "/content/test_images"
os.makedirs(save_dir, exist_ok=True)

# ✅ List of image URLs
image_urls = [
    "https://cdn.agclassroom.org/media/uploads/tomato3.jpg",
    "http://media1.popsugar-assets.com/files/2013/12/20/030/n/1922195/a0c1d85ae04404b6_shutterstock_116285338.jpg.xxxlarge_2x/i/Cabbage.jpg",
    "https://upload.wikimedia.org/wikipedia/commons/thumb/2/25/Onion_on_White.JPG/1200px-Onion_on_White.JPG",
    "https://www.thespruceeats.com/thmb/bPWzkkBT5vbXlwrRQXNrPBUWxwY=/2336x0/filters:no_upscale():max_bytes(150000):strip_icc()/fresh-ginger-whole-and-chopped-on-rustic-wood-surface-162711007-ed63813f6b2a4acbaeb3e5ff1da17fa2.jpg"
    # 👉 Add more URLs here
]

# ✅ Download each image and give it a unique name
for i, url in enumerate(image_urls):
    filename = f"image_{i+1}.jpg"
    save_path = os.path.join(save_dir, filename)
    try:
        urllib.request.urlretrieve(url, save_path)
        print(f"Downloaded {filename}")
    except Exception as e:
        print(f"Failed to download {url}: {e}")
```
